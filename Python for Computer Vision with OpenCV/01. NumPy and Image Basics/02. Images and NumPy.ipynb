```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from PIL import Image
```

```python
pic = Image.open('../img/cute-puppy.jpg')
```

```python
type(pic)
```

```python
pic_arr = np.asanyarray(pic)
```

```python
type(pic_arr)
```

```python
# show 3 for RGB array
pic_arr.shape
```

```python
plt.imshow(pic)
```

```python
# Image array got copied
pic_red = pic_arr.copy()
plt.imshow(pic_red)
```

```python
plt.imshow(pic_red[:,:,0],cmap = 'gray')
```

```python
# To just get the Green color from the picture

# in this copied whole image and set red and blue value to 0

pic_green = pic_arr.copy()
pic_green[:,:,0] = 0
pic_green[:,:,2] = 0
plt.imshow(pic_green)
```

```python
# cmap will map gray color where value is green 
plt.imshow(pic_green[:,:,1],cmap = 'gray')
```

```python
# as for red color it was set 0 so it showing gray
plt.imshow(pic_green[:,:,0],cmap = 'gray')
```

```python
# To just get the blue color from the picture

# in this method just copied blue colour from the pic_arr
pic_blue = pic_arr[:,:,2]
pic_blue.shape
```

```python
plt.imshow(pic_blue)
```
